# Visualize relationships between features and the target variable.

```python
# Set plot style @sabinvankathmandu
sns.set_theme(style="whitegrid")
```

```python
# Histogram of features @sabinvankathmandu
df.hist(figsize=(12, 10), bins=20, edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
# Boxplots for Each Column for outliers detections @sabinvankathmandu

# Setting the figure size @sabinvankathmandu
plt.figure(figsize=(15, 10))

# create a boxplot by looping for each columns @sabinvankathmandu
for i, column in enumerate(df.columns, 1):
    plt.subplot(4, 4, i)  # adjusting rows & cols as needed (4x4 for 16 features)
    sns.boxplot(y=df[column])
    plt.title(f"Boxplot of {column}")

plt.tight_layout()  # layout adujst to prevent overlapping @sabinvankathmandu
plt.show()
```

```python
# Heat mAp or Correlation matrix @sabinvankathmandu

plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Matrix")
plt.show()
```

### Distribution of Target Variables (MEDV)

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["medv"], bins=30, kde=True)
plt.title("Distribution of MEDV (House Prices)")
plt.xlabel("MEDV (in $1000s)")
plt.ylabel("Frequency")
plt.show()
```

### Relationship Between MEDV (House Price) & Key Features

```python
# top correlated features with MEDV @sabinvankathmandu
correlated_features = df.corr()["medv"].abs().sort_values(ascending=False)
print("Top Correlated Features with MEDV:\n", correlated_features)

# scatter plots for key factors
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
sns.scatterplot(x=df["rm"], y=df["medv"], ax=axes[0, 0]).set_title("Rooms vs Price")
sns.scatterplot(x=df["lstat"], y=df["medv"], ax=axes[0, 1]).set_title("LSTAT vs Price")
sns.scatterplot(x=df["ptratio"], y=df["medv"], ax=axes[1, 0]).set_title("PTRATIO vs Price")
sns.scatterplot(x=df["crim"], y=df["medv"], ax=axes[1, 1]).set_title("Crime Rate vs Price")

plt.tight_layout()
plt.show()
```
